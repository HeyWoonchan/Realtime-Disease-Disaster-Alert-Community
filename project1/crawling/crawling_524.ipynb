WHO 뉴스 크롤링

```python
import requests
from bs4 import BeautifulSoup
import sqlite3
from urllib.parse import urljoin

url = "https://www.who.int/emergencies/disease-outbreak-news"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

news_articles = soup.find_all('a', class_='sf-list-vertical__item')

# 데이터베이스 연결
conn = sqlite3.connect('ExternalNews.db')
cursor = conn.cursor()

# ExternalNews 테이블 생성
create_table_query = '''
CREATE TABLE IF NOT EXISTS ExternalNews (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    link TEXT
)
'''
cursor.execute(create_table_query)

# 링크 저장 및 데이터 삽입
for article in news_articles:
    title = article.text.strip()
    link = article.get('href').strip()
    link_text = urljoin(url, link)
    insert_query = 'INSERT INTO ExternalNews (title, link) VALUES (?, ?)'
    cursor.execute(insert_query, (title, link_text))
    print(f"제목: {title}")
    print(f"링크: {link_text}")
    print()

# 변경사항 저장
conn.commit()
# 데이터베이스 연결 종료
conn.close()
```

자연재난 행동요령 크롤링

```python
import requests
import sqlite3
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url = "https://www.safekorea.go.kr/idsiSFK/neo/sfk/cs/contents/prevent/prevent01.html?menuSeq=126"

response = requests.get(url)
html_content = response.text

soup = BeautifulSoup(response.content, 'html.parser')

inner_helps = soup.find_all('a', class_='fireGuide_link')

conn = sqlite3.connect('InternalNature.db')
cursor = conn.cursor()
create_table_query = '''
CREATE TABLE IF NOT EXISTS InternalNature (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    link TEXT
)
''' 
cursor.execute(create_table_query)

base_url = "https://www.safekorea.go.kr/idsiSFK/neo/sfk/cs/contents/prevent/"

for article in inner_helps:
    title = article.text
    onclick_attr = article.get('onclick')
    if onclick_attr:
        link_start = onclick_attr.find("'") + 1
        link_end = onclick_attr.rfind("'")
        link = onclick_attr[link_start:link_end]
        absolute_link = urljoin(base_url, link)
        insert_query = 'INSERT INTO InternalNature (title, link) VALUES (?, ?)'
        cursor.execute(insert_query, (title, absolute_link))
        print(f"제목: {title}")
        print(f"링크: {absolute_link}")
        print()
    else:
        href = article.get('href')
        absolute_href = urljoin(base_url, href)
        insert_query = 'INSERT INTO InternalNature (title, link) VALUES (?, ?)'
        cursor.execute(insert_query, (title, absolute_href))
        print(f"제목: {title}")
        print(f"링크: {absolute_href}")
        print()

# 변경사항 저장
conn.commit()
# 데이터베이스 연결 종료
conn.close()
```
